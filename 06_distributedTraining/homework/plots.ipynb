```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('metrics.dat')
plt.plot(data[:,0])
plt.plot(data[:,2])
plt.legend(['train accuracy','validation accuracy'])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.savefig('accuracy_1.pdf') 
plt.show()
```
